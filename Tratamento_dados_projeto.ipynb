```python
import pandas as pd
```

```python
df = pd.read_csv('/content/fkmrodado.xlsx - fKmRodado.csv')
display(df)
```

```python
#renomeando coluna tirando espaços e acentos

df.rename(columns={'Mês':'mes','ID Veiculo':'id_veiculo','Km percorridos':'km_percorridos', 'Gasto com Combustível':'valor_combustivel', 'Manut.':'Manutencao', 'Custos Fixos': 'custos_fixos'
                   },inplace=True)
df.head(1)
```

```python
df.info()
```

```python
df['valor_combustivel'] = df['valor_combustivel'].astype('string')
```

```python
df.info()
```

```python
#verificando duplicatas

df.duplicated().any()
```

```python
df['data_corrigida'] = pd.to_datetime(df['mes'], format='%m/%d/%Y')

# Verificando o DataFrame com a data corrigida
print(df)
```

```python
#dropando coluna com data errada

df = df.drop(columns=['mes'])
```

```python
# Construir colunas de ano, mes e dia para facilitar o agrupamento posterior
df['ano'] = df['data_corrigida'].dt.year
df['mes'] = df['data_corrigida'].dt.month
df['dia'] = df['data_corrigida'].dt.day

print(df)
```

```python
display(df)
```

```python
#salvando em CSV
df.to_csv('dados_corrigidos_kmrodados.csv', index=False)
```

#Tratamento base frete

link csv frete:

https://docs.google.com/spreadsheets/d/1DEcQ0I21-JNYF8tP2bZeDkLQBQLSwJ7B/edit?usp=sharing&ouid=106799284520595176840&rtpof=true&sd=true

```python
frete = pd.read_csv('/content/ffrete.xlsx - fFrete.csv')
display(frete)
```

```python
frete.info()
```

```python
#renomeando coluna tirando espaços e acentos

frete.rename(columns={'Data':'data','ID Cliente':'ID_Cliente','ID Veiculo':'id_veiculo', 'Numero Documento Fiscal':'notafiscal', 'Viagem':'viagem', 'Valor do Frete Líquido':'valor_frete_liquido',
                       'Peso (KG)':'kg', 'Valor da Mercadoria':'valor_mercadoria'
                   },inplace=True)
frete.head(1)
```

```python
frete['valor_mercadoria'] = frete['valor_mercadoria'].str.replace(',', '.', regex=False)
```

```python
frete['valor_mercadoria'] = pd.to_numeric(frete['valor_mercadoria'], errors='coerce')
```

```python
frete['valor_mercadoria'] = frete['valor_mercadoria'].round(2)
```

```python
print(frete)
```

```python
#tratando coluna viagem

frete[['data_viagem', 'placa_veiculo']] = frete['viagem'].str.split('-', n=1, expand=True)

# Verificando o resultado
print(frete)
```

```python
#dropando coluna viagem

frete = frete.drop(columns=['viagem'])
```

```python
frete['date_viagem'] = pd.to_datetime(frete['data_viagem'], format='%d/%m/%Y')

# Verificando o DataFrame com a data corrigida
print(frete)
```

```python
display(frete)
```

```python
#dropando coluna viagem

frete = frete.drop(columns=['data_viagem'])
```

```python
frete.info()
```

```python
frete['data_corrigida'] = pd.to_datetime(frete['data'], format='%m/%d/%Y')

# Verificando o DataFrame com a data corrigida
print(frete)
```

```python
#dropando coluna data

frete = frete.drop(columns=['data'])
```

```python
frete.info()
```

```python
frete['notafiscal'] = frete['notafiscal'].astype('string')
frete['placa_veiculo'] = frete['placa_veiculo'].astype('string')
```

```python
# Passo 1: Substituir as vírgulas por pontos para valores decimais
frete['kg'] = frete['kg'].str.replace(',', '.', regex=False)

# Passo 2: Converter a coluna para tipo numérico (float), ignorando erros
frete['kg'] = pd.to_numeric(frete['kg'], errors='coerce')
```

```python
frete.info()
print(frete)
```

```python
# Passo 1: Substituir as vírgulas por pontos para valores decimais
frete['valor_frete_liquido'] = frete['valor_frete_liquido'].str.replace(',', '.', regex=False)

# Passo 2: Converter a coluna para tipo numérico (float), ignorando erros
frete['valor_frete_liquido'] = pd.to_numeric(frete['valor_frete_liquido'], errors='coerce')

frete['valor_frete_liquido'] = frete['valor_frete_liquido'].round(2)
```

```python
frete.info()
print(frete)
```

```python
frete.duplicated().any()
```

```python
#salvando em CSV
frete.to_csv('dados_corrigidos_frete.csv', index=False)
```
